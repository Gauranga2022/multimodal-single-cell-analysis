## Data Preprocessing

#### Normalization

```python
columns_to_drop = ['cell_id', 'day', 'donor', 'technology', 'cell_type']
X_train = X_train.drop(columns=columns_to_drop, axis=1)
```

```python
X_train.head()
```

```python
X_train1 = pd.DataFrame(X_train)
scaler = StandardScaler()
X_train1 = scaler.fit_transform(X_train1)
X_train1 = pd.DataFrame(X_train1)
X_train1.head()
```

#### Label Encoding

```python
y_train.head()
```

```python
le = LabelEncoder()
y_train = le.fit_transform(y_train.iloc[:,0])
le.classes_
```

```python
y_train = pd.DataFrame(y_train)
y_train.head()
```

#### PCA

```python
df = pd.DataFrame(X_train1)
pca = PCA(n_components=10000)
transformed_data = pca.fit_transform(X_train1)
```

```python
data = pd.DataFrame(transformed_data)
```

```python
data.head()
```

```python
data.info()
```

```python
#pca=65
sns.set(style="whitegrid")
fig, ax = plt.subplots()
sns.scatterplot(data=data, x=6, y=9, hue=y_train.iloc[:,0], palette="pastel")
```

```python
pca = PCA(n_components=3)
pca_result = pca.fit_transform(data.values)

pca_df = pd.DataFrame(pca_result, columns=['PC1', 'PC2', 'PC3'], index=data.index)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(pca_df['PC1'], pca_df['PC2'], pca_df['PC3'], c=y_train)

ax.set_xlabel('PC1')
ax.set_ylabel('PC2')
ax.set_zlabel('PC3')
ax.set_title('PCA 3D Scatter Plot')

plt.show()
```
